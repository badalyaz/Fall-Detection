```python
from utils import FeatureExtractor, KeyPoints    #Import FeatureExtractor and KeyPoints class
```

```python
featureextractor = FeatureExtractor()    #Create the featureextractor instance of FeatureExtractor class
```

```python
vid = './/TestData//Fall3.mp4'
costmethod = 'Division'
```

```python
def getCosts(vid, costmethod):
    """
    Function getCosts
    Used for getting the cost list of the entire video
    Takes as input the video path and cost method
    Returns the list of the costs for the video
    
    """
    cost = featureextractor.processVideo(vid, costmethod)
    
    return cost
```

```python
def plot(cost, fall_start, fall_end):
    
    """
    Function plot
    Used for plotting the cost list
    Takes as input the cost, starting frame of the fall and ending frame
    Returns the plot
    
    """
    
    plot = plt.figure(figsize=(5,5))

    plt.plot(cost, label = 'cost')
    
    plt.title('Cost method is: ' + costmethod)
    
    plt.axhline(y = featureextractor.threshold, label = 'Threshold', color='black')

    plt.axvspan(fall_start, fall_end, alpha=0.25, color='red', label = 'Fall Frames')   

    plt.legend(loc="upper right")
    
    plt.show()
```

```python
cost = getCosts(vid, costmethod)
```

```python
plot(cost, 17-6, 25-6)
```
